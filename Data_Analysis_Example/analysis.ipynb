### Imports

```python
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder


import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import joblib
```

```python
df = pd.read_csv("/Users/homefolder/Desktop/Machine Learning Basics Repo/Data_Analysis_Example/data/student_performance.csv")

df.head()
```

```python
df.info()
```

```python
# Notice in the df, there are 'object' data types
# These object datatypes cannot be used in out model
# Therefore we can either not include them OR convert them

le = LabelEncoder()

df['ParentalSupport'] = le.fit_transform(df['ParentalSupport'])
```

```python
df.info()
```

```python
# Create variable for the features 
# that you want to train with and the target (what you want to predict)

features = df[['AttendanceRate','StudyHoursPerWeek','PreviousGrade','ExtracurricularActivities','ParentalSupport']]

target = df['FinalGrade']
```

```python
# Perform train_test_split to improve model
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size=0.2, random_state=42)
```

```python
# initialize basic machine learning models
ml_linear = LinearRegression()
ml_rf = RandomForestRegressor()
ml_svr = svm.SVC()

# Train the models
ml_linear.fit(X_train, y_train)
ml_rf.fit(X_train, y_train)
ml_svr.fit(X_train, y_train)
```

```python
'AttendanceRate','StudyHoursPerWeek','PreviousGrade','ExtracurricularActivities','ParentalSupport'
input1 = 50
input2 = 5
input3 = 2
input4 = 0
input5 = 0

complete_input = [[input1, input2, input3, input4, input5]]

result = ml_linear.predict(complete_input)
result2 = ml_rf.predict(complete_input)
result3 = ml_svr.predict(complete_input)

returnToOrginal = le.inverse_transform(result3)

print(returnToOrginal)
```

```python
# Let us measure how well the models are
prediction = ml_linear.predict(X_test)

print(prediction)
```

```python
mae = mean_absolute_error(y_test, prediction)
print(mae)
```
